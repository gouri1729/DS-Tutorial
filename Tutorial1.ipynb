```python
path="/content/drive/My Drive/ds/Advertising.csv"
df=pd.read_csv(path)
```

```python
x=df[['TV','radio','newspaper']]
y=df['sales']
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
for i in x:
  plt.figure(figsize=(7,3))
  plt.title(f'{i} vs Sales',fontsize=14)
  plt.xlabel(f'{i}',fontsize=15)
  plt.ylabel('Sales',fontsize=15)
  plt.scatter(df[i],df['sales'],alpha=0.7,color='orange')
  plt.show()
```

```python
from sklearn.linear_model import LinearRegression

for i in x:

    model=LinearRegression()
    model.fit(x[[i]],y)
    pred_y=model.predict(x[[i]])
    plt.scatter(df[i],pred_y,alpha=0.6,color='yellow')
    plt.plot(df[[i]], pred_y, color='blue', label='Regression Line')
    plt.show()

    R_sq=model.score(x[[i]],y)
    print(R_sq)
```
