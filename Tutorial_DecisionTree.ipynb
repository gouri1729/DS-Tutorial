```python
import numpy as np 
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report
```

```python
from sklearn.datasets import load_breast_cancer

dataset = load_breast_cancer()
```

```python
data = pd.DataFrame(dataset.data,columns=dataset.feature_names)
data['target'] = dataset.target
```

```python
data.describe()
```

```python
data.shape
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(data[dataset.feature_names],data['target'],random_state=42)
```

```python
len(X_train), len(X_test), len(Y_train) , len(Y_test)
```

```python
model = DecisionTreeClassifier(criterion="entropy",splitter="best",random_state=42)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
```

```python
accuracy
```

```python
class_rpt = classification_report(Y_test,Y_pred)
```

```python
print(class_rpt)
```
